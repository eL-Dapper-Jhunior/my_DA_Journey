```python
yahoo['Daily Return'] = (yahoo['Close'] - yahoo['Open']) / yahoo['Open']
# Or use pct_change from the close price
# yahoo['Daily Return'] = yahoo['Close'].pct_change()
yahoo
```

```python
# 10-day Moving Average
yahoo['MA_10'] = yahoo['Close'].rolling(window=10).mean()

# 50-day and 200-day MAs for long-term trend analysis
yahoo['MA_50'] = yahoo['Close'].rolling(window=50).mean()
yahoo['MA_200'] = yahoo['Close'].rolling(window=200).mean()
yahoo
```

```python
# 10-day Volatility (standard deviation of daily returns)
yahoo['Volatility_10'] = yahoo['Daily Return'].rolling(window=10).std()
yahoo
```

```python
# 10-day EMA
yahoo['EMA_10'] = yahoo['Close'].ewm(span=10, adjust=False).mean()
yahoo
```

```python
# Calculate price changes
delta = yahoo['Close'].diff()

# Separate gains and losses
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)

# Average gains/losses over 14 periods
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

# Calculate RS and RSI
rs = avg_gain / avg_loss
yahoo['RSI_14'] = 100 - (100 / (1 + rs))

yahoo
```

```python
# Calculate EMAs
ema_12 = yahoo['Close'].ewm(span=12, adjust=False).mean()
ema_26 = yahoo['Close'].ewm(span=26, adjust=False).mean()

# MACD line
yahoo['MACD'] = ema_12 - ema_26

# Signal line (9-day EMA of MACD)
yahoo['Signal_Line'] = yahoo['MACD'].ewm(span=9, adjust=False).mean()
yahoo
```

```python
# 20-day moving average and std deviation
ma_20 = yahoo['Close'].rolling(window=20).mean()
std_20 = yahoo['Close'].rolling(window=20).std()

# Upper and lower bands
yahoo['Bollinger_Upper'] = ma_20 + (2 * std_20)
yahoo['Bollinger_Lower'] = ma_20 - (2 * std_20)
yahoo
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Set date as index
yahoo['Date'] = pd.to_datetime(yahoo['Date'])
yahoo.set_index('Date', inplace=True)

# Choose the column to forecast
ts = yahoo['Close']

# Build ARIMA model (p, d, q) - start with (5,1,0) as a test
model = ARIMA(ts, order=(5, 1, 0))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=30)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(ts, label='Historical')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.title('30-Day Forecast')
plt.show()
```

```python
# Reset index to turn 'Date' back into a column
yahoo = yahoo.reset_index()

# Now rename for Prophet
prophet_df = yahoo[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})

# Sort by date (optional but good practice)
prophet_df = prophet_df.sort_values('ds')
```

```python
from prophet import Prophet

model = Prophet()
model.fit(prophet_df)
```

```python
# Create future dates
future = model.make_future_dataframe(periods=30)

# Forecast
forecast = model.predict(future)

# Show key results
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
import matplotlib.pyplot as plt

# Plot forecast
fig = model.plot(forecast)

# Annotate the last 30 forecast points
for i in range(-30, 0):
    date = forecast['ds'].iloc[i]
    yhat = forecast['yhat'].iloc[i]
    plt.text(date, yhat, f'{yhat:.2f}', fontsize=8, color='red', rotation=45, ha='left', va='bottom')

plt.title('Prophet 30-Day Stock Price Forecast')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
